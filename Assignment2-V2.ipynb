Question 5) Create a new column named as 'happiness_degree' which labels value of 'ladder'
*very happy if ladder < 30
*happy if 30 <= ladder < 60
*moderately happy if 60 <= ladder < 90
*unhappy if 90 <= ladder < 120
*very unhappy if ladder >= 120
Print the top 10 records updated dataframe (10 points)

```python
#codes
happiness_df.loc[happiness_df.Ladder < 30, "happiness_degree"] = "Very Happy"
happiness_df.loc[(happiness_df.Ladder >= 30) & (happiness_df.Ladder<60), "happiness_degree"] = "Happy"
happiness_df.loc[(happiness_df.Ladder >= 60) & (happiness_df.Ladder<90), "happiness_degree"] = "moderately happy"
happiness_df.loc[(happiness_df.Ladder >= 90) & (happiness_df.Ladder<120), "happiness_degree"] = "unhappy"
happiness_df.loc[(happiness_df.Ladder >= 120) , "happiness_degree"] = "very unhappy"
display(happiness_df[:10])
    
    
```

Question 10) Draw random samples 30%, 40%, 50%, 60% of the data with replacement and investigate how the correlation of sample to the whole data changes with respect to sample size. (15 points)

```python
#codes
thirty= happiness_df.sample(frac=0.30,replace=True,random_state=1)
fourty = happiness_df.sample(frac=0.40,replace=True,random_state=1)
fifty = happiness_df.sample(frac=0.50,replace=True,random_state=1)
sixty = happiness_df.sample(frac=0.60,replace=True,random_state=1)
print("Thirty Percent sample correlation")
display(thirty.corr())
print("Fourty Percent sample correlation")
display(fourty.corr())
print("Fifty Percent sample correlation")          ##WHEN SAMPLE FRACTION GETS BIGGER 
display(fifty.corr())                                ## IN GENERAL VALUES GETS CLOSER TO DEFAULT CORR VALUES
print("Sixty Percent sample correlation")
display(sixty.corr())
display(happiness_df.corr())
```
